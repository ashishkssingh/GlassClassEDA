```python
df = pd.read_csv('winequality-red.csv')
```

## How features are related to quality

### Fixed Acidity and Volatile acidity to quality

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='fixed acidity',y='volatile acidity',hue='quality',data=df)
```

 - Lower volatile acidity gives higher quality
 - Fixed acidity does not affect quality much
 - Volatile acidity plays bigger role in understanding quality

### Citric Acid and Residual Sugar to quality

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='citric acid',y='residual sugar',hue='quality',data=df)
```

 - Residual sugar does not help understanding quality, its always low with few outliers.
 - Citric acid helps us understand higher quality wines.

### Chlorides and Sulphur affecting the quality

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='chlorides',y='total sulfur dioxide',hue='quality',data=df)
```

 - Chlorides don't vary quality of wine.
 - Total Sulfur dioxide does not affects quality.

### Density and Sulphates affecting the quality

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x='density',y='sulphates',hue='quality',data=df)
```

 - Sulphates helps us in identifying good quality wines.
 - Density does not affect as much.

### Alcohol quanity affecting wine quality

```python
plt.figure(figsize=(15,10))
sns.barplot(y='alcohol',x='quality',data=df)
```

 - Higher level of alcohols constitutes for higher quality wine. 

## Check of Distribution of quality affecting features

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
sns.violinplot(x='quality',y='volatile acidity',data=df)

plt.subplot(2,2,2)
sns.violinplot(x='quality',y='citric acid',data=df)

plt.subplot(2,2,3)
sns.violinplot(x='quality',y='sulphates',data=df)

plt.subplot(2,2,4)
sns.violinplot(x='quality',y='alcohol',data=df)
```

## Confirm our observations using correlation and pair plot.

### Visualizing pair plot

```python
# Select features
features = ['volatile acidity','citric acid', 'sulphates', 'alcohol', 'quality']

plt.figure(figsize=(10,8))
sns.pairplot(df[features], diag_kind='hist', hue='quality')
```

### Calculate correlation

```python
corr = df.corr()
corr
```

### Plotting correlation heat map

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True)
```

**Confirmed Observation using Heat map**
1. Volatile acidity plays bigger role in understanding quality.
2. Citric acid helps us understand higher quality wines.
3. Sulphates helps us in identifying good quality wines.
4. Higher level of alcohols constitutes for higher quality wine.
